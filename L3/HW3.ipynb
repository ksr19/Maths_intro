__Задание 3.__ Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

__Решение:__ Напишем соответствующие функции, которые будут принимать необходимые параметры, а затем строить график:

```python
def circle(r):
    # каноническое уравнение окружности является вырожденным случаем уравнения эллипса: x^2/r^2 + y^2/r^2 = 1
    # тогда y^2 = r^2 - x^2
    x = np.linspace(-r, r, 1000)
    y = np.sqrt(r**2 - x**2)
    plt.plot(x, y)
    plt.plot(x, -y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('equal')
    plt.title(f'Окружность с радиусом {r}')
    plt.grid()
    plt.show()
    return None

def ellipse(a, b):
    # каноническое уравнение эллипса: x^2/a^2 + y^2/b^2 = 1
    # тогда y^2 = (a^2*b^2 - b^2*x^2)/a^2
    x = np.linspace(-a, a, 1000)
    y = np.sqrt(((a**2)*(b**2) - (b**2)*x**2) / a**2)
    plt.plot(x, y)
    plt.plot(x, -y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('equal')
    plt.title(f'Эллипс с полуосями {a} и {b}')
    plt.grid()
    plt.show()
    return None

def hyperbola(a, b):
    # каноническое уравнение гиперболы: x^2/a^2 - y^2/b^2 = 1
    # тогда y^2 = (b^2*x^2 - a^2*b^2)/a^2
    x1 = np.linspace(-a-20, -a, 1000)
    x2 = np.linspace(a, a+20, 1000)
    y1 = np.sqrt(((b**2)*x1**2 - (a**2)*(b**2)) / a**2)
    y2 = np.sqrt(((b**2)*x2**2 - (a**2)*(b**2)) / a**2)
    plt.plot(x1, y1)
    plt.plot(x2, y2)
    plt.plot(x1, -y1)
    plt.plot(x2, -y2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('equal')
    plt.title(f'Гипербола с полуосями {a} и {b}')
    plt.grid()
    plt.show()
    return None
```

```python
circle(r=10)
```

```python
ellipse(a=10, b=5)
```

```python
hyperbola(a=10, b=5)
```

2) Пусть задана плоскость $A_{1}x+B_{1}y+C_{1}z+D_{1}=0$ и прямая $\frac{x-x_{1}}{x_{2}-x_{1}}=\frac{y-y_{1}}{y_{2}-y_{1}}=\frac{z-z_{1}}{z_{2}-z_{1}}$.<br>
Как узнать, принадлежит прямая плоскости или нет?

__Решение:__ <br>
1) Искомая плоскость параллельна заданной, а значит вектор $(A, B, C)$ также является для нее нормальным. Кроме того, она проходит через начало координат. Значит, уравнение искомой плоскости имеет вид:
### $$A(x-0) + B(y-0) + C(z-0) = 0 \Rightarrow Ax + By + Cz=0$$

Визуализируем плоскости, предположив, что $A = 2; B = 5; C=1; D = 13$.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = figure(figsize=(12, 8))
ax = Axes3D(fig)
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(X, Y)
Z1 = -13 - 2*X - 5*Y
Z2 = -2*X - 5*Y
ax.plot_surface(X, Y, Z1, cmap=cm.gray)
ax.plot_surface(X, Y, Z2, cmap=cm.spring)
plt.show()
```

2) Прямая принадлежит плоскости, если обе точки, задающие прямую, $(x_{1}, y_{1}, z_{1})$ и $(x_{2}, y_{2}, z_{2})$  находятся на плоскости, то есть следующие выражения являются справедливыми $A_{1}x_{1}+B_{1}y_{1}+C_{1}z_{1}+D_{1}=0$ и $A_{1}x_{2}+B_{1}y_{2}+C_{1}z_{2}+D_{1}=0$. Или переписав иначе, когда справедливо $A_{1}(x_{2}-x_{1})+B_{1}(y_{2}-y_{1})+C_{1}(z_{2}-z_{1})=0$, то есть когда скалярное произведение направляющего вектора прямой и вектора, перпендикулярного плоскости, равно нулю.

Допустим, что точки, определяющие прямую, это $(5; 0; 1)$ и $(2; 3; 0)$, а уравнение плоскости $2x + 5y + z + 13 = 0$. Тогда выражение для проверки, записанное ранее, примет следующий вид:
### $$2\cdot(2-5) + 5\cdot(3-0) + 1\cdot(0-1) \neq 0$$
То есть прямая, проходящая через эти точки, не принадлежит заданной плоскости.

Рассмотрим теперь случай, когда точки, определяющие прямую, это $(-2; -1; -4)$ и $(0; -3; 2)$, а уравнение плоскости остается неизменным: $2x + 5y + z + 13 = 0$. Тогда выражение будет выглядеть следующим образом:
### $$2\cdot(0+2) + 5\cdot(-3+1) + 1\cdot(2+4) = 0$$
А значит прямая, проходящая через эти точки, принадлежит заданной плоскости.

__Задание 2.__ <br>
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

__Решение:__ <br>
Проведем доказательство для случая двумерного пространства, при необходимости оно может быть обобщено для случая $n-$мерного пространства. <br>
Допустим у нас есть две точки $k = (x_{1}, y_{1})$ и $m = (x_{2}, y_{2})$. Расстояние между этими точками определяется по следующей формуле: $\sqrt{(x_{2}-x_{1})^2 + (y_{2}-y_{1})^2}$. <br>
После ортогонального преобразования координаты точек изменились и теперь записываются следующим образом:
$k' = (a_{11}x_{1}+a_{12}y_{1}+a_{13}, a_{21}x_{1}+a_{22}y_{1}+a_{23})$ и $m' = (a_{11}x_{2}+a_{12}y_{2}+a_{13}, a_{21}x_{2}+a_{22}y_{2}+a_{23})$. Запишем формулу по которой можно рассчитать расстояние между точками $k'$ и $m'$:
$\sqrt{(a_{11}x_{2}+a_{12}y_{2}+a_{13}-a_{11}x_{1}-a_{12}y_{1}-a_{13})^2 + (a_{21}x_{2}+a_{22}y_{2}+a_{23}-a_{21}x_{1}-a_{22}y_{1}-a_{23})^2}$. <br>
После сокращения свободных членов и группировки формула принимает следующий вид:
$\sqrt{(a_{11}(x_{2}-x_{1})+a_{12}(y_{2}-y_{1}))^2 + (a_{21}(x_{2}-x_{1})+a_{22}(y_{2}-y_{1}))^2}$. Раскрывая квадраты, получаем следующее:
$\sqrt{a_{11}^2(x_{2}-x_{1})^2+a_{12}^2(y_{2}-y_{1})^2 + 2a_{12}a_{11}(x_{2}-x_{1})(y_{2}-y_{1}) + a_{21}^2(x_{2}-x_{1})^2+a_{22}^2(y_{2}-y_{1})^2 + 2a_{22}a_{21}(x_{2}-x_{1})(y_{2}-y_{1})}$. <br>
В силу свойств ортогонального преобразования: 
1. $a_{11}^2 + a_{21}^2=1$;
2. $a_{12}^2 + a_{22}^2=1$;
3. $a_{11}a_{12} + a_{21}a_{22}=0$.


С учетом этого записанное выше расстояние между точками $k'$ и $m'$ может быть переписано как $\sqrt{(x_{2}-x_{1})^2 + (y_{2}-y_{1})^2}$, что в точности совпадает с расстоянием между точками $k$ и $m$. Значит, при ортогональном преобразовании расстояние между точками сохраняется, что и требовалось доказать.

```python
# (x1, y1) и (x2, y2) - координаты концов отрезка в декартовой системе

def polar_line_segment(x1, y1, x2, y2):
    x = np.linspace(np.arccos(x1/np.sqrt(x1**2 + y1**2)), np.arccos(x2/np.sqrt(x2**2 + y2**2)), 1000)
    y = (-y1 + x1*(y2-y1)/(x2-x1)) / (np.cos(x) * (y2-y1)/(x2-x1) - np.sin(x))
    plt.polar(x, y)
    plt.ylim(0, np.max(y) + 1)
    plt.show()
    
polar_line_segment(x1=0, y1=2, x2=2, y2=4)
polar_line_segment(x1=-2, y1=2, x2=-9, y2=2)
```

__Задание 4.__ Решите систему уравнений и неравенств: <br>

  e^x + x(1-y) = 1, \\

2)
### $$\left\{
\begin{array}{ll}
  y = x^2 - 1, \\
  e^x + x(1-y) - 1 > 0.
\end{array} \right. $$

__Решение:__ <br>
1) Для начала построим графики, чтобы оценить количество решений.

```python
x1 = np.linspace(-6, -1e-15, 201)
x2 = np.linspace(1e-15, 6, 201)
x = np.append(x1, x2)
plt.figure(figsize=(12, 8))
plt.plot(x1, (np.exp(x1) - 1 + x1)/x1, 'red')
plt.plot(x2, (np.exp(x2) - 1 + x2)/x2, 'red')
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Как видим, система имеет как минимум три решения. Кроме того в силу построения присутствует неопределенность в нуле, которая может оказаться дополнительным, четвертым решением.

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y-x**2+1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(equations, (-5, 5))
x2, y2 = fsolve(equations, (1, 1))
x3, y3 = fsolve(equations, (3, 3))
x4, y4 = fsolve(equations, (6, 6))

print(f'Решение 1 - {round(x1, 6), round(y1, 6)}.')
print(f'Решение 2 - {round(x2, 6), round(y2, 6)}.')
print(f'Решение 3 - {round(x3, 6), round(y3, 6)}.')
print(f'Решение 4 - {round(x4, 6), round(y4, 6)}.')
```
