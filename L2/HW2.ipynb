1. Напишите уравнение параболы, проходящей через три точки (x,y): (1,2), (3,10), (5,1).

__Решение:__ В общем виде уравнение параболы имеет вид: $y = ax^2 + bx + c$. Подставляя координат точек в это выражение, получаем следующую систему уравнений:
### $$\left\{
\begin{array}{ll}
  a + b + c = 2, \\
  9a + 3b + c = 10,  \\
  25a + 5b + c = 1.
\end{array} \right.$$
Выразим из первого уравнения $c = 2 - a - b$ и подставим во второе уравнение системы. Получим следующее:
### $$9a + 3b + 2 -a - b = 10 \Rightarrow 8a + 2b = 8 \Rightarrow b = 4 - 4a$$
Возвращаясь к выражению $c$ через $a$ и $b$ и с учетом найденной выше связи между $b$ и $a$:
### $$ c = 2 - a - b = 2 - a - 4 + 4a = -2 + 3a$$
Подставим выражения $c$ и $b$ в последнее уравнение системы:
### $$25a + 20 - 20a - 2 + 3a = 1 \Rightarrow 8a = -17 \Rightarrow a = -2.125$$
Тогда уравнение параболы выглядит следующим образом:
### $$y = -2.125x^2 + 12.5x -8.375$$

2. Известно, что свежий огурец на 99% состоит из воды. Месяц назад взвесили мешок со свежими огурцами. Получилось, что мешок весил ровно 100 кг. Мешок убрали, а через месяц снова взвесили. Огурцы за это время усохли, и теперь вода составляет уже только 98% их веса. Сколько теперь (в кг) весят огурцы?

__Решение:__
При первом взвешивании вода составляла 99% массы огурцов, то есть с учетом изначального веса 100кг, вода весила 99кг, значит масса полезного вещества составляет 1кг, заметим, что она не изменяется с течением времени. Через месяц вода составляет 98% веса. Можно составить следующее уравнение:
### $$B = 0.98 (B + 1) \Rightarrow 0.02B = 0.98 \Rightarrow B = 49$$
___Значит вес огурцов через месяц 49 + 1 = 50кг.___

3. Определение логарифма. Решить уравнения:


1. $2^x = 256$
2. $2^x = 300$
3. $\log_{8}{2^{8x-4}}=4$
4. $3^{\log_{9}{(5x-5)}}=5$
5. $x^{\log_{3}{x}+1} = 9$

__Решение:__
1. $x = 8$;
2. $x = \log_{2}{300}$;
3. Заметим, что $8 = 2^3$, тогда пользуюсь свойствами логарифмов получаем следующее уравнение:
$\frac{8x-4}{3} = 4 \Rightarrow x = 2;$
4. Прологарифмируем обе части уравнения по основанию 3, тогда
### $$\log_{9}{(5x-5)} = \log_{3}{5} \Rightarrow \log_{3}{(5x-5)}=2\log_{3}{5}\Rightarrow 5x - 5 = 25 \Rightarrow x=6;$$
5. Прологарифмируем обе части уравнения по основанию 3, тогда
### $$\log_{3}{x^{\log_{3}{x}+1}}=2 \Rightarrow (\log_{3}{x} + 1)\cdot\log_{3}{x} = 2$$
Произведем следующую замену переменной: $t = \log_{3}{x}$. Тогда получаем следующее квадратное уравнение:
### $$t^2 + t - 2 = 0$$
Оно имеет следующие решения: $t_{1,2} = \frac{-1 \pm \sqrt{1 - 4\cdot(-2)}}{2} \Rightarrow t_1 = -2; t_2 = 1.$
Производя обратную замену получаем итоговое решение уравнения: 
### $$\log_{3}{x} = -2 \Rightarrow x_1 = \frac{1}{9}$$
### $$\log_{3}{x} = 1 \Rightarrow x_2 = 3$$
