2. 1) Решите систему уравнений: 
### $$\left\{
\begin{array}{ll}
  3x - 2y + 5z = 7, \\
  7x + 4y - 8z = 3,  \\
  5x - 3y - 4z = -12.
\end{array} \right.$$

Линейная или нелинейная это система? А каждое уравнение по отдельности?

___Решение___: <br>
Все уравнения линейные и сама система тоже линейная.
### $$\left\{
\begin{array}{ll}
  3x - 2y + 5z = 7, \\
  7x + 4y - 8z = 3,  \\
  5x - 3y - 4z = -12.
\end{array} \right.$$
Умножим второе уравнение на 3 и вычтем из него первое, умноженное на 7, а из третьего уравнения, умноженного на 3, вычтем первое уравнение, умноженное на 5. Тогда,
### $$\left\{
\begin{array}{ll}
  3x - 2y + 5z = 7, \\
  21x + 12y - 24z - 21x + 14y - 35z = 9 - 49,  \\
  15x - 9y - 12z - 15x + 10y -25z= -36 - 35.
\end{array} \right. \Rightarrow \left\{
\begin{array}{ll}
  3x - 2y + 5z = 7, \\
  26y - 59z = -40,  \\
  y - 37z= -71.
\end{array} \right.$$
Значит, $ y = 37z - 71$. Подставим это выражение во второе уравнение и получим:
### $$26\cdot 37z - 26\cdot71 - 59z = -40 \Rightarrow 903z = 1806 \Rightarrow z = 2.$$
Тогда, $ y = 37\cdot2 - 71  = 3$. Найдем $x$, подставив вычисленные $y$ и $z$ в первое уравнение системы:
### $$3x-2\cdot3 + 5\cdot2 = 7 \Rightarrow 3x = 3 \Rightarrow x = 1.$$
Итоговое решение системы (единственное):
### $$\left\{
\begin{array}{ll}
  x=1, \\
  y=3,  \\
  z=2.
\end{array} \right.$$

2) Решите систему уравнений: 
### $$\left\{
\begin{array}{ll}
  x^2+yx-9 = 0, \\
  x - \frac{y}{5} = 0. 
\end{array} \right.$$

Линейная или нелинейная это система? А каждое уравнение по отдельности?

___Решение___:<br>
Первое уравнение нелинейное, второе линейное. Значит, система уравнений нелинейная.
### $$\left\{
\begin{array}{ll}
  x^2+yx-9 = 0, \\
  x - \frac{y}{5} = 0. 
\end{array} \right. \Rightarrow 
\left\{
\begin{array}{ll}
  x^2+yx-9 = 0, \\
  y = 5x. 
\end{array} \right.$$
Подставляя $y$, выраженный через $x$, в первое уравнение системы, получаем следующее:
### $$6x^2 = 9 \Rightarrow  x = \pm\sqrt{\frac{3}{2}}.$$
Это означает, что система имеет два решения. Первое - $(\sqrt{\frac{3}{2}}; 5\sqrt{\frac{3}{2}})$, второе - $(-\sqrt{\frac{3}{2}}; -5\sqrt{\frac{3}{2}})$.

3. Площадь пола прямоугольной комнаты равна 48 м2, а его периметр равен 28 м. Найдите длину и ширину комнаты.

___Решение___:
Пусть $x$ - длина комната, тогда ширину можно выразить через периметр: $\frac{28-2x}{2} = 14 -x.$ <br>
Площадь прямоугольной комнаты равна произведению длины на ширину, а значит имеем уравнение:
### $$14x - x^2 = 48 \Rightarrow x^2-14x + 48 = 0.$$
Найдем дискриминант:
### $$D = 14^2 - 4\cdot1\cdot48 = 4.$$
Уравнение имеет два решения, но исходя из логики длина должна быть больше ширины, а значит нас интересует только одно:
### $$x=\frac{14 + \sqrt{4}}{2} = 8.$$
Итого, __длина комнаты - 8м, а ширина - 6м__.

4. Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 1500)
k1 = 1
k2 = 3
plt.figure(figsize=(12, 8))
plt.plot(x, np.cos(k1*x), color='b', label='cos(x)')
plt.plot(x, np.cos(k2*x), color='r', label=f'cos({k2}x)')
plt.legend()
plt.grid()
```
