__Задание 1.__ Решите уравнение $\frac{\sin{x}}{x} = 0$.

__Решение:__ <br>
Прежде всего обратим внимание на ОДЗ: $x \neq 0$.
Левая часть обращается в ноль, когда числитель равен нулю, а знаменатель отличен от нуля, то есть:
### $$\left\{
\begin{array}{ll}
  \sin{x} = 0, \\
  x \neq 0.
\end{array} \right.$$
Так как синус периодическая функция, то решением данной системы будет множество значений. В общем виде решение исходного уравнения может быть записано следующим образом:
### $$x = \pi n, n \in \mathbb Z, n \neq 0$$

```python
x = np.linspace(-10, 10, 10000)
y = np.sin(x)

plt.figure(figsize=(12, 10))
plt.plot(x, y)
plt.plot(x, [0] * len(x), 'r')

for i in range(1, int(max(x)/np.pi)):
    plt.axvline(np.pi * i, linestyle=':')
    plt.axvline(-np.pi * i, linestyle=':')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

__Задание 2.__ Даны три прямые $y_{1} = k_{1}x + b_{1}$, $y_{2} = k_{2}x + b_{2}$ и $y_{3} = k_{3}x + b_{3}$. Как узнать, пересекаются ли они в одной точке или нет?

__Решение:__ <br>
Одним из вариантов является графическое решение, которое заключается в том, чтобы построить эти прямые и посмотреть пересекаются ли они и, если да, то каким образом. <br> 
Существует также аналитическое решение. Если три прямые пересекаются в одной точке, то существует $(x_{0}; y_{0})$ для которой справедливо:
### $$y_{0} = k_{1}x_{0} + b_{1}=k_{2}x_{0} + b_{2} = k_{3}x_{0} + b_{3}$$
Отсюда можно выразить $x_{0}$, например, из условия пересечения первых двух прямых:
### $$k_{1}x_{0} + b_{1}=k_{2}x_{0} + b_{2} \Rightarrow x_{0} = \frac{b_{2} - b_{1}}{k_{1} - k_{2}}$$
Заметим при этом, что знаменатель не равен нулю, так как иначе это бы означало параллельность прямых, что делает невозможным их пересечение в точке. Подставим выражение для $x_{0}$ в условие пересечения второй и третьей прямых:
### $$k_{2}\frac{b_{2} - b_{1}}{k_{1} - k_{2}} + b_{2} = k_{3}\frac{b_{2} - b_{1}}{k_{1} - k_{2}} + b_{3}$$
Тогда, приведя обе части к общему знаменателю, получаем:
### $$k_{2}b_{2}-k_{2}b_{1} + k_{1}b_{2} - k_{2}b_{2} = k_{3}b_{2} - k_{3}b_{1} + k_{1}b_{3} -  k_{2}b_{3}$$
Перегруппируем слагаемые и получаем выражение:
### $$b_{1}(k_{2}-k_{3}) + b_{2}(k_{3} - k_{1}) + b_{3}(k_{1} - k_{2}) = 0$$
Если полученное условие выполняется и при этом прямые различны, то есть ни одна из прямых не совпадает полностью с другой, то они пересекаются в одной точке. Если условие не выполняется, то заданные три прямые не пересекаются в одной точке.

__Задание 3.__ На листе тетради «в линейку» (расстояние между линиями равно а) лежит игла (длиной b). Координаты нижней точки иглы (х,у), игла лежит под углом alfa. Пересекает ли игла линию или нет?

__Решение:__ <br>
Из условия задачи не до конца понятно каким образом измеряется угол, под которым лежит игла, в частности может ли он быть тупым. Поэтому будем исходить из того, что угол изменяется от 0 до 180 градусов. Можно представить, что игла играет роль гипотенузы в прямоугольном треугольнике (за исключением вырожденных случаев когда $\alpha = 0^{\circ}, 90^{\circ}, 180^{\circ}$). Для того, чтобы понять пересекает ли она линию, нам понадобится длина противоположного катета, которая может быть вычислена по формуле $b \cdot \sin{\alpha}$. Тогда исходя из того, что существует такое $ k \in \mathbb Z$, что $ka \leq y < (k+1)a$, можно сформулировать условие, при выполнении которого игла будет пересекать линию:
### $$(k+1)a - y < b\sin{\alpha}$$
### $$\sin{\alpha} > \frac{(k+1)a - y}{b}$$
В противном случае игла не пересекает линию.

__Задание 4.__ Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра $a$: 
$\sin{(ax)}=0$ при условии: $0.01<a<0.02, 100<x<500$.
Т.е. надо найти решение $x$ как функцию параметра $а$ - построить график $x=x(a)$.
Если численным методом не получается найти все ветви решения $x(a)$, то отыщите хотя бы одну.

__Решение:__ <br>
Как было уже отмечено в задании 1, решением уравнения $\sin{(ax)}=0$ является множество $\pi n, n \in \mathbb Z$. Значит, 
### $$ax = \pi n, n \in \mathbb Z \Rightarrow x = \frac{\pi n}{a}, n \in \mathbb Z$$
При этом из условия задачи известно, что $0.01<a<0.02, 100<х<500$. Заметим, что $\frac{\pi}{0.01} \approx 314.1593$, а $\frac{\pi}{0.02} \approx 157.0796$. Тогда в зависимости от значения параметра $a$ возможны следующие решения заданного уравнения, удовлетворяющие ограничению по значению $x$:
1) $a \in (0.01; \frac{\pi}{250}]$: $ x = \frac{\pi}{a}$;<br>
2) $a \in (\frac{\pi}{250}; \frac{3\pi}{500}]$: $ x = \frac{\pi n}{a}, n = 1, 2$; <br>
3) $a \in (\frac{3\pi}{500}; 0.02)$: $ x = \frac{\pi n}{a}, n = 1, 2, 3$.

```python
a1 = np.linspace(0.01, 0.02, 10000)[1:-1]
a2 = np.linspace(np.pi/250, 0.02, 10000)[1:-1]
a3 = np.linspace(3*np.pi/500, 0.02, 10000)[1:-1]
x1 = np.pi/a1
x2 = 2*np.pi/a2
x3 = 3*np.pi/a3

plt.figure(figsize=(12, 10))
plt.plot(a1, x1)
plt.plot(a2, x2)
plt.plot(a3, x3)
plt.axvline(np.pi/250, linestyle=':')
plt.axvline(3*np.pi/500, linestyle=':')
plt.grid()
plt.ylim(70, 520)
plt.xlabel('a')
plt.ylabel('x')
plt.title('Значения x, являющиеся решением уравнения sin(ax) = 0, в зависимости от параметра a')
plt.show()
```

Данный график позволяет определить значения $x$, которые являются решениями уравнения $\sin{(ax)} = 0$, при конкретном заданном значении параметра $a$.

```python
def find_roots(a):
    # Исходим из предположения, что передается числовое значение параметра
    if a < 0.01 or a > 0.02:
        return f'Значение параметра выходит за заданные пределы'
    elif a <= np.pi/250:
        x = [np.pi / a]
    elif a <= 3*np.pi/500:
        x = [np.pi / a, 2*np.pi / a]
    else:
        x = [np.pi / a, 2*np.pi / a, 3*np.pi / a]
    
    return x

print(f'a = 0.05: {find_roots(0.05)}')
print(f'a = 0.012: {find_roots(0.012)}')
print(f'a = 0.014: {find_roots(0.014)}')
print(f'a = 0.019: {find_roots(0.019)}')
```

### $$(y^2 - 2y + 2) - 2x - 7=0 \Rightarrow (y-2)^2 - 2x - 7 = 0 \Rightarrow (y-2)^2 = 2x+7$$
Полученное уравнение отличается от канонического вида параболы $y^2 = 2px$ только наличием сдвига. Таким образом, данное уравнение порождает __параболу__.
