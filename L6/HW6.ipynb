__Задание 1 (теорема сложения).__
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

__Решение:__ <br>
Выпадение 2 очков и выпадение 5 очков при подбрасывании игральной кости являются несовместными событиями, а значит искомая вероятность:
### $$P(A = 2 | A =5 ) = P(A = 2) + P(A = 5) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3}.$$

__Задание 2 (теорема умножения).__ Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

__Решение:__ <br>
Так как события несовместные, то искомая вероятность будет равна произведению вероятности:
### $$P(A = 2; B = 5) = P(A = 2) \cdot P(B = 5) = \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}.$$

__Задание 3.__
Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости. Обратите внимание на порядок выпадения костей!

__Решение:__ <br>
Здесь возможны два варианты, каждый из которых имеет одинаковую вероятность, найденную в предыдущем пункте:
1) При первом подбрасывании выпадает двойка, а при втором - пятерка; <br>
2) При первом подбрасывании выпадает пятерка, а при втором - двойка. 

Тогда искомая вероятность может быть рассчитана следующим образом:
### $$ P = P(A = 2; B = 5) + P(A = 5; B = 2) = 2\cdot\frac{1}{6} \cdot \frac{1}{6} = \frac{1}{18}.$$

__Задание 4 (Геометрическая вероятность + интервалы).__ На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

__Решение:__ <br>
Если обозначим координату точки $A$ за $0$,  а координату точки $B$ за $20$, то для того, чтобы удовлетворять условию точка $C$ должна принадлежать отрезку $[5; 9]$. Первая граница получается из условия, что точка $C$ должна быть не дальше $15$ см от точки $B$. Вторая граница получается из условия, что точка $C$ должна быть не дальше $9$ см от точки $A$. Длина отрезка - $4$ см, а длина отрезка $AB$ - 20 см. Значит вероятность того, что точка $C$ находится на расстоянии не более $9$ см от точки $А$ и не более $15$ см от точки $В$:
### $$P_{C} = \frac{4}{20} = 0.2$$

__Задание 7.__ Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью  все грани этого куба будут белыми?

__Решение:__ <br>
Заметим, что после разрезания у нас получилось 26 кубиков, у которых окрашена в белый цвет хотя бы одна грань, и один черный кубик, который находился в центре исходного куба. Дополнительно отметим, что каждая грань состоит из 9 кубиков (для соседних граней часть из них является общими). <br>
Для того, чтобы собрать куб, у которого все грани будут белыми необходимо, чтобы одновременно выполнялся ряд условий:
1. в центре каждой грани находятся кубики, у которых только одна грань белая. Всего таких кубиков столько же, сколько граней в исходном большом кубе - 6; <br>
2. кубики, которые находятся в центрах ребер, должны иметь две соседние наружные грани белые, так как ребра являются общими для двух граней. Каждая грань содержит 4 таких кубика - по одному на ребро, однако помним, что ребра общие для соседних граней. Всего таких кубиков столько же, сколько ребер в исходном большом кубе - 12; <br>
3. кубики, которые находятся в вершинах куба, должны иметь 3 соседние белые грани, так как эти кубики будут общими сразу для трех граней. Каждая грань содержит 4 таких кубика, которые находятся в углах. В силу того, что они являются общими сразу для трех граней, количество таких кубиков равняется количеству вершин куба - 8; <br>
4. в центре куба должен находиться полностью черный кубик из центра исходного куба. <br>


Перейдем к сборке кубика и подсчету вероятности по пунктам:
1. Вероятность, с которой мы можем выбрать все 6 кубиков, у которых только одна грань белая, из общего набора 27 кубиков определяется по формуле: 
### $$\frac{C^6_6}{C^6_{27}} = \frac{6!21!}{27!}$$
Кроме того нужно, чтобы каждый из кубиков лег единственной белой гранью (из 6) наружу. Вероятность этого - $\big(\frac{1}{6}\big)^6$. Таким образом, общая вероятность данного пункта:
### $$\frac{6!21!}{27!}\cdot\frac{1}{2^6\cdot3^6}$$
2. После предыдущего пункта у нас остался 21 кубик, из которого нужно выбрать все 12 кубиков, у которых две грани белые. Вероятность этого:
### $$\frac{C^{12}_{12}}{C^{12}_{21}} = \frac{12!9!}{21!}$$
Здесь также нужно рассчитать вероятность того, что каждый из выбранных кубиков правильно расположен. Можно начать по аналогии с предыдущим пунктом, рассчитав вероятность одной конкретной выбранной белой грани выпасть наружу - $\frac{1}{6}$. Однако важно, чтобы одна из 4 соседних граней кубика, которая окажется на соседней грани большого куба, тоже была белая. Далее отметим, что для нас важен только цвет грани (аналогично с задачей 3, где нам не важен порядок выпадения), а значит вероятность определяется следующим образом:
### $$\big(2!\cdot\frac{1}{6}\cdot\frac{1}{4}\big)^{12}$$
Итоговая вероятность данного пункта:
### $$\frac{12!9!}{21!}\cdot\frac{1}{2^{24}\cdot3^{12}}$$
3. У нас осталось 9 кубиков, из которых мы должны выбрать все 8, обладающие тремя белыми гранями. Вероятность такого выбора:
### $$\frac{C^{8}_{8}}{C^{8}_{9}} = \frac{1!8!}{9!}$$
Снова определим вероятность того, что все кубики расположились при случайной сборке правильно. В качестве базы возьмем предыдущий пункт: то есть сначала выбираем одну белую грань, считаем вероятность попадания ее наружу, затем одна выбранная грань из 4 соседних должна тоже оказаться снаружи, наконец одна из двух граней, которые являются соседними с первыми двумя выбранными тоже должна быть белой и тоже должна оказаться снаружи - на третьей грани большого куба. Снова для нас важен только цвет, а не порядок, тогда вероятность:
### $$\big(3!\cdot\frac{1}{6}\cdot\frac{1}{4}\cdot\frac{1}{2}\big)^{8}$$
Итоговая вероятность данного пункта:
### $$\frac{1!8!}{9!}\cdot\frac{1}{2^{24}}$$
4. У нас остался 1 кубик, который полностью черный и будет находиться в центре большого куба, поэтому нам не важно какие грани находятся выше/ниже. Значит, итоговая вероятность данного пункта - $1$.


Соберем вероятности всех пунктов. Напомним, что условия должны выполняться одновременно, а значит, вероятность того, что все грани собранного куба будут белыми:
### $$\frac{6!21!}{27!}\cdot\frac{1}{2^6\cdot3^6}\cdot\frac{12!9!}{21!}\cdot\frac{1}{2^{24}\cdot3^{12}}\cdot\frac{1!8!}{9!}\cdot\frac{1}{2^{24}}\cdot1 = \frac{6!8!12!}{27!}\cdot\frac{1}{2^{54}\cdot3^{18}}$$
Рассчитаем числовое значение с помощью NumPy.

```python
np.math.factorial(6) * np.math.factorial(12) * np.math.factorial(8) / (np.math.factorial(27) * 2**54 * 3**18)
```
